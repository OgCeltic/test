```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    data = pd.read_csv('C:\\Users\\30128\\Downloads\\titanic.csv')

    data.dropna(inplace=True)

    data = data[data['Age'] <= 80]

    plt.figure(figsize=(10, 6))
    plt.scatter(data['Age'], data['Fare'])
    plt.xlabel('年龄')
    plt.ylabel('票价')
    plt.title('Age vs Fare')
    plt.show()

    plt.figure(figsize=(10, 6))
    data['Survived'].value_counts().sort_index(ascending=True).plot(kind='line',
                                                                      marker='o',
                                                                      linestyle='-')
    plt.title('存活人数趋势')
    plt.xlabel('类别')
    plt.ylabel('存活人数')
    plt.show()

except FileNotFoundError:
    print("文件不存在。")
    
except Exception as e:
    print(f"读取文件时发生错误：{e}")
```
