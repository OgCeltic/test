```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    data = pd.read_csv('C:\\Users\\30128\\Downloads\\titanic.csv')

    data.dropna(inplace=True)

    data = data[data['Age'] <= 80]

    plt.figure(figsize=(10, 6))
    plt.scatter(data['Age'], data['Fare'])
    plt.xlabel('')
    plt.ylabel('')
    plt.title('Age vs Fare')
    plt.show()

    plt.figure(figsize=(10, 6))
    data['Survived'].value_counts().sort_index(ascending=True).plot(kind='line',
                                                                      marker='o',
                                                                      linestyle='-')
    plt.title('')
    plt.xlabel('')
    plt.ylabel('')
    plt.show()

except FileNotFoundError:
    print("")
    
except Exception as e:
    print(f"{e}")
```
